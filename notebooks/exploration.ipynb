```python
import requests
import polars as pl
```

```python
target_year = 2026
target_month = 2
file_name = f'ist-daten-v2-{target_year}-{target_month:02d}.zip'
```

```python
def get_source_url():
    buckets = [
    f'https://archive.opentransportdata.swiss/istdaten/{target_year}/',
    'https://archive.opentransportdata.swiss/istdaten/2025/']
    
    for bucket in buckets:
        url = bucket + file_name
        response = requests.head(url)
        if response.ok:
            return url
```
